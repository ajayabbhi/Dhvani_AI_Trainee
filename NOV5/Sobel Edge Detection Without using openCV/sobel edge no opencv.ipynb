# Sobel Edge Detection not using openCv


### Dependencies

```python
from scipy import ndimage
from matplotlib import pyplot as plt
import numpy as np  
```

### Read image w/o opencv

```python
import imageio.v2 as imageio
image = imageio.imread('Rose.jpg', mode='F')
```

### Sobel_x,Sobel_y matrix

```python
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])  
```

Gradient in X,Y Direction

```python
gx = ndimage.convolve(image,sobel_x)
gy = ndimage.convolve(image,sobel_y)
```

### Magnitude

```python
mag = np.sqrt(gx**2 + gy **2)
mag = mag.astype(np.uint8) #convert into 8bit image for display
```

### Display

```python
plt.figure(figsize=(12, 12))
plt.subplot(1,2,1),plt.title('Original Image') , plt.imshow(image, cmap='gray')
plt.subplot(1,2,2),plt.title('Sobel Edge Detection') , plt.imshow(mag, cmap='gray')
plt.axis('off')             
```
