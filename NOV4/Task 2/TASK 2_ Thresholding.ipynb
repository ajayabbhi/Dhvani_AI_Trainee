# TASK 2_ THRESHOLDING

### DEPENDENCIES

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Convert an image to grayscale and apply a fixed threshold.

```python
#convert to grayscale
img = cv2.imread('DRAGON.JPEG', cv2.IMREAD_GRAYSCALE)

# Apply a fixed threshold
a = int(input("Enter threshold value (0-255): "))
_, th_fixed = cv2.threshold(img, a, 255, cv2.THRESH_BINARY)

# Show results
plt.subplot(1,2,1), plt.title("Original"), plt.imshow(img,cmap = 'gray')
plt.subplot(1,2,2), plt.title(f"Fixed Threshold {a}"), plt.imshow(th_fixed, cmap = 'gray')
plt.axis('off')
plt.show()
```

## Adaptive Thresholding_used when uneven lighting on image

```python
# Adaptive mean
th_mean = cv2.adaptiveThreshold(img, 255,
                                cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY, 11, 2)

# Adaptive Gaussian
th_gaussian = cv2.adaptiveThreshold(img, 255,
                                    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                    cv2.THRESH_BINARY, 11, 2)

cv2.imwrite('mean.png',th_gaussian)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.title("Adaptive Mean"), plt.imshow(th_mean ,cmap = 'gray') #color map = gray

# why cmap used ?
# By default, Matplotlib(plt.imshow) assumes your image is in RGB format 

plt.subplot(1,2,2), plt.title("Adaptive Gaussian"), plt.imshow(th_gaussian, cmap='gray')
plt.show()
```

## Otsus thresholding (automatic threshold selection).

```python
import cv2
from matplotlib import pyplot as plt

# Read image in grayscale
img = cv2.imread('DRAGON.JPEG', cv2.IMREAD_GRAYSCALE)

# Apply Otsus thresholding
ret, otsu_thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print(f"Otsu's automatically selected threshold: {ret}")

# Show results
plt.subplot(1,2,1), plt.title("Original"), plt.imshow(img,cmap = 'gray')
plt.subplot(1,2,2), plt.title("Otsu's Threshold"), plt.imshow(otsu_thresh, cmap = 'gray')
plt.axis('off')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Threshold in different planes - RGB, HSV

## RGB

### Fixed/Global Thresholding.

```python
import cv2
import matplotlib.pyplot as plt

# Read the image and convert to RGB (OpenCV reads as BGR by default)
img = cv2.imread('ball.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split into R, G, B channels
R, G, B = cv2.split(img)

Red = int(input("Enter threshold value for Red channel (0-255): "))
Green = int(input("Enter threshold value for Green channel (0-255): "))
Blue = int(input("Enter threshold value for Blue channel (0-255): "))

# Apply fixed threshold (value = 127)
Red,th_R = cv2.threshold(R, Red, 255, cv2.THRESH_BINARY)
Green, th_G = cv2.threshold(G, Green, 255, cv2.THRESH_BINARY)
Blue, th_B = cv2.threshold(B, Blue, 255, cv2.THRESH_BINARY)

print(f"User-defined thresholds - R: {Red}, G: {Green}, B: {Blue}") 

# Plot results
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1); plt.imshow(img); plt.title("Original RGB"); plt.axis('off')
plt.subplot(2, 2, 2); plt.imshow(th_R, cmap='gray'); plt.title("Fixed Threshold - R"); plt.axis('off')
plt.subplot(2, 2, 3); plt.imshow(th_G, cmap='gray'); plt.title("Fixed Threshold - G"); plt.axis('off')
plt.subplot(2, 2, 4); plt.imshow(th_B, cmap='gray'); plt.title("Fixed Threshold - B"); plt.axis('off')
plt.tight_layout()
plt.show()
```

### Otsus Thresholding

```python
import cv2
import matplotlib.pyplot as plt

# Read the image and convert to RGB
img = cv2.imread('ball.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split into R, G, B channels
R, G, B = cv2.split(img)

# Apply Otsu's thresholding (automatic)
r_value, th_R = cv2.threshold(R, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
g_value, th_G = cv2.threshold(G, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
b_value, th_B = cv2.threshold(B, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print(f"Otsu's automatically selected thresholds - R: {r_value}, G: {g_value}, B: {b_value}")

# Plot results
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1); plt.imshow(img); plt.title("Original RGB"); plt.axis('off')
plt.subplot(2, 2, 2); plt.imshow(th_R, cmap='gray'); plt.title("Otsu Threshold - R"); plt.axis('off')
plt.subplot(2, 2, 3); plt.imshow(th_G, cmap='gray'); plt.title("Otsu Threshold - G"); plt.axis('off')
plt.subplot(2, 2, 4); plt.imshow(th_B, cmap='gray'); plt.title("Otsu Threshold - B"); plt.axis('off')
plt.tight_layout()
plt.show()
```

## HSV

### Fixed (Global) Thresholding

```python
import cv2
import matplotlib.pyplot as plt

# Read image and convert from BGR  HSV
img = cv2.imread('ball.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Split HSV channels
H, S, V = cv2.split(img)

H_value = int(input("Enter threshold value for Hue channel (0-179): "))
S_value = int(input("Enter threshold value for Saturation channel (0-255): "))
V_value = int(input("Enter threshold value for Value channel (0-255): "))

# Apply fixed (global) threshold at user-defined values
HUE, th_H = cv2.threshold(H, H_value, 255, cv2.THRESH_BINARY)
SAT, th_S = cv2.threshold(S, S_value, 255, cv2.THRESH_BINARY)
VAL, th_V = cv2.threshold(V, V_value, 255, cv2.THRESH_BINARY)

print(f"User-defined thresholds - H: {H_value}, S: {S_value}, V: {V_value}")

# Plot results
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB)); plt.title("Original (HSVRGB)"); plt.axis('off')
plt.subplot(2, 2, 2); plt.imshow(th_H, cmap='gray'); plt.title("Fixed Threshold - Hue"); plt.axis('off')
plt.subplot(2, 2, 3); plt.imshow(th_S, cmap='gray'); plt.title("Fixed Threshold - Saturation"); plt.axis('off')
plt.subplot(2, 2, 4); plt.imshow(th_V, cmap='gray'); plt.title("Fixed Threshold - Value"); plt.axis('off')
plt.tight_layout()
plt.show()
```

## Visualize histograms before and after thresholding.

### Fixed (Global) Thresholding

```python
import cv2
import matplotlib.pyplot as plt

# Load a simple grayscale image
img = cv2.imread(cv2.samples.findFile('Sudoku.png'), cv2.IMREAD_GRAYSCALE)

# Apply fixed threshold
_, th_fixed = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Plot image and histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), bins=256, range=(0, 256), color='gray')
plt.title('Histogram Before Thresholding')

plt.subplot(2, 2, 3)
plt.imshow(th_fixed, cmap='gray')
plt.title('Fixed Thresholded Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(th_fixed.ravel(), bins=256, range=(0, 256), color='red')
plt.title('Histogram After Thresholding')

plt.tight_layout()
plt.show()
```


### Adaptive Thresholding

```python
import cv2
import matplotlib.pyplot as plt

# Load grayscale image
img = cv2.imread(cv2.samples.findFile('Sudoku.png'), cv2.IMREAD_GRAYSCALE)

# Apply Adaptive Gaussian Thresholding
th_adapt = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY, 11, 2)

# Plot image and histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), bins=256, range=(0, 256), color='gray')
plt.title('Histogram Before Thresholding')

plt.subplot(2, 2, 3)
plt.imshow(th_adapt, cmap='gray')
plt.title('Adaptive Thresholded Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(th_adapt.ravel(), bins=256, range=(0, 256), color='blue')
plt.title('Histogram After Thresholding')

plt.tight_layout()
plt.show()
```

### Otsus Thresholding

```python
import cv2
import matplotlib.pyplot as plt

# Load grayscale image
img = cv2.imread(cv2.samples.findFile('Sudoku.png'), cv2.IMREAD_GRAYSCALE)

# Apply Otsu's thresholding
ret, th_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(f"Otsus computed threshold value: {ret}")

# Plot image and histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), bins=256, range=(0, 256), color='gray')
plt.axvline(ret, color='r', linestyle='dashed', linewidth=2)
plt.title('Histogram Before Thresholding (Otsu line in red)')

plt.subplot(2, 2, 3)
plt.imshow(th_otsu, cmap='gray')
plt.title('Otsu Thresholded Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(th_otsu.ravel(), bins=256, range=(0, 256), color='green')
plt.title('Histogram After Thresholding')

plt.tight_layout()
plt.show()
```
