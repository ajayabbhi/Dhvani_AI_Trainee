# Task 1 - Sobel Edges Detection


## Apply Sobel filter in x and y directions separately.


### Dependencies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read image
img = cv2.imread('chess.jpg')
#convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)
plt.title('Original Grayscale Image')
plt.imshow(gray, cmap='gray')
plt.axis('off')    
```

### Apply Sobel filter in x(vertical edges) and y(horizontal edges)

```python
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)  # Sobel filter in x direction
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)  # Sobel filter in y direction

# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1),plt.title('Sobel X'),plt.imshow(sobel_x, cmap='gray')

plt.subplot(1, 2, 2),plt.title('Sobel Y'),plt.imshow(sobel_y, cmap='gray')

plt.axis('off')
```

## Combine them to visualize edge magnitude.

```python
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
magnitude_8bit = np.uint8(np.clip(magnitude, 0, 255)) # Convert to 8-bit image

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1),plt.title('Edge Magnitude 64 float'),plt.imshow(magnitude, cmap='gray')
plt.subplot(1, 2, 2),plt.title('Edge Magnitude 8 bits'),plt.imshow(magnitude_8bit, cmap='gray')
plt.axis('off')
```

## Try on different images (grayscale and color).

### Color

```python
img_color = cv2.imread('chess.jpg')
plt.figure(figsize=(12, 6))
img_color_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.subplot(1, 1, 1),plt.title('Color Image - Original'),plt.imshow(img_color_rgb)
plt.axis('off')
```

### Apply Sobel filter in x,y direction and Combine them to visualize edge magnitude in color image

```python
sobel_x_color = cv2.Sobel(img_color, cv2.CV_64F, 1, 0, ksize=5)  # Sobel filter in x direction
sobel_y_color = cv2.Sobel(img_color, cv2.CV_64F, 0, 1, ksize=5)  # Sobel filter in y direction
magnitude_color = np.sqrt(sobel_x_color**2 + sobel_y_color**2)
magnitude_color_8bit = np.uint8(np.clip(magnitude_color, 0, 255)) # Convert to 8-bit image

plt.Figure(figsize=(18, 18))
plt.subplot(2, 2, 1),plt.title('sobel_x'),plt.imshow(sobel_x_color, cmap='gray')
plt.subplot(2, 2, 2),plt.title('sobel_y'),plt.imshow(sobel_y_color, cmap='gray')
plt.subplot(2, 2, 3),plt.title('sobel_magnitude'),plt.imshow(magnitude_color, cmap='gray')
plt.subplot(2, 2, 4),plt.title('sobel_magnitude_8bit'),plt.imshow(magnitude_color_8bit, cmap='gray')
plt.tight_layout()
plt.axis('off')
```
