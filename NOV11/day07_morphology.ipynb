# Morphological Operations in OpenCV

## 2. Dilation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)

kernel = np.ones((13,13),np.uint8)

dilation = cv2.dilate(thresh,kernel,10)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,2,2),plt.imshow(dilation,cmap ='gray'),plt.title('Dilation') #object enlargement
plt.tight_layout()
```

## 3. Opening 

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)


_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,13))

opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

kernel = np.ones((13,13),np.uint8)
manual_way  = cv2.dilate(erosion,kernel,1)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,3,2),plt.imshow(opening,cmap ='gray'),plt.title('Opening')
plt.subplot(1,3,3),plt.imshow(manual_way,cmap ='gray'),plt.title('Manual Way Opening') # Erosion followed by Dilation
plt.tight_layout()
```

##  4. Closing

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)

kernel = np.ones((13,13),np.uint8)
manual_way = cv2.erode(dilation,kernel,1)


kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,13))

closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,3,2),plt.imshow(closing,cmap ='gray'),plt.title('Closing')
plt.subplot(1,3,3),plt.imshow(manual_way,cmap ='gray'),plt.title('Manual Way Closing') # Dilation followed by Erosion
plt.tight_layout()
```

## 6. TopHat

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)


th = np.subtract(thresh,opening)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,13))

Tophat = cv2.morphologyEx(thresh, cv2.MORPH_TOPHAT, kernel)

plt.figure(figsize=(8,8))

plt.subplot(2,2,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(2,2,2),plt.imshow(opening,cmap ='gray'),plt.title('Opening')
plt.subplot(2,2,3),plt.imshow(Tophat,cmap ='gray'),plt.title('Tophat')
plt.subplot(2,2,4),plt.imshow(th,cmap ='gray'),plt.title('Tophat by subtraction') # image minus opening.
plt.tight_layout()
```

## 7. BlackHat

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)

BH = np.subtract(closing,thresh)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,13))

Blackhat = cv2.morphologyEx(thresh, cv2.MORPH_BLACKHAT, kernel)


plt.figure(figsize=(8,8))

plt.subplot(2,2,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(2,2,2),plt.imshow(closing,cmap ='gray'),plt.title('Closing')
plt.subplot(2,2,3),plt.imshow(Blackhat,cmap ='gray'),plt.title('BlackHat')
plt.subplot(2,2,4),plt.imshow(BH,cmap ='gray'),plt.title('BlackHat by subtraction') # closing minus Image 
plt.tight_layout()
```

## 8.Hit Or Miss

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)


kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,13))

HITORMISS = cv2.morphologyEx(thresh, cv2.MORPH_HITMISS, kernel)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,2,2),plt.imshow(HITORMISS,cmap ='gray'),plt.title('Hit-or-Miss')
plt.tight_layout()
```

## 9. THINNING

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)


kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))

HITORMISS = cv2.morphologyEx(thresh, cv2.MORPH_HITMISS, kernel)

thin = np.subtract(thresh, HITORMISS)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,3,2),plt.imshow(HITORMISS,cmap ='gray'),plt.title('Hit-or-Miss')
plt.subplot(1,3,3),plt.imshow(thin,cmap ='gray'),plt.title('Thinning')           #image minus Hit-or-Miss
plt.tight_layout()
```

## 10. Thickening

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))

HITORMISS = cv2.morphologyEx(thresh, cv2.MORPH_HITMISS, kernel)
Thickening = cv2.bitwise_or(thresh, HITORMISS)
plt.figure(figsize=(10,5))


plt.subplot(1,3,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,3,2),plt.imshow(HITORMISS,cmap ='gray'),plt.title('HIT-or-Miss')
plt.subplot(1,3,3),plt.imshow(Thickening,cmap ='gray'),plt.title('Thickening')             #image (OR) Hit-or-Miss
plt.tight_layout()
```

## 11. Skeletonization (Medial Axis Transformation)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

path = 'A.jpg'
img = cv2.imread(path , cv2.IMREAD_GRAYSCALE)

_,thresh = cv2.threshold(img,128,255,cv2.THRESH_BINARY_INV)


skeleton = cv2.distanceTransform(thresh, cv2.DIST_L2, 5)

plt.figure(figsize=(10,5))


plt.subplot(1,3,1),plt.imshow(thresh, cmap='gray'),plt.title('Original')
plt.subplot(1,3,2),plt.imshow(skeleton,cmap ='gray'),plt.title('Skeleton')
plt.tight_layout()
```
