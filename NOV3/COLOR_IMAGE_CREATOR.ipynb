# COLOR_IMAGE_CREATOR


### Dependencies

```python
import cv2
import numpy as np
```

### RGB VALUES


```python
#BGR values between 0–255
b,g,r = 255,255,255 #WHY BGR...OPENCV SEES AS BGR
width, height = 400, 300  # image size
```

### Create solid color image with np.full



```python
# np.full() → directly creates all-color base

color = (b, g, r)
image = np.full((height, width, 3), color, dtype=np.uint8)
```

### Create solid color image with np.Zeros



```python
# np.zeros() → creates all-zero (black) base
import numpy as np
import cv2
image = np.zeros((1000, 1000, 3),np.uint8)
image[:] = (10, 50, 70)

cv2.imshow("Color Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Save the image 

```python
filename = f"color_{r}_{g}_{b}.png"
cv2.imwrite(filename, image)

print(f"✅ Image saved as {filename}")
print(image.shape)
```

## Blend 2image



### Read both color images

```python

img1 = cv2.imread("color_255_0_255.png")
img2 = cv2.imread("color_0_255_0.png")

```

### Resize both to same size

```python
#img_shape = (height, width, channels)


height = min(img1.shape[0], img2.shape[0])
width = min(img1.shape[1], img2.shape[1])
img1 = cv2.resize(img1, (width, height))  #Order of arguments:OpenCV, it’s (width, height) 
img2 = cv2.resize(img2, (width, height))

```

### Mix both images 

```python

# alpha and beta control the blending ratio
alpha = 0.5  # weight for first image
beta = 0.5   # weight for second image

blended = cv2.addWeighted(img1, alpha, img2, beta,100)
```

### Save the result


```python

cv2.imwrite("blended_image.png", blended)
print("✅ Saved: blended_image.png")
```

## BGR TO GREYSCALING

```python
image_path = 'color_120_200_150.png'

img = cv2.imread(image_path)

imag = blended

gray = cv2.cvtColor(imag, cv2.COLOR_BGR2GRAY)

# Save the grayscale image
cv2.imwrite("grayscale_image.png", gray)

print("✅ Saved: grayscale_image.png")
```

## GRAY TO BGR


```python
rgb_from_gray = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

# Save or display
cv2.imwrite("gray_to_rgb.png", rgb_from_gray)
print("✅ Saved: gray_to_rgb.png")
```
